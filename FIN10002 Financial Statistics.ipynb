```python
df = pd.read_csv ('random_sampling_fixed_data.csv')
```

```python
df
```

```python
Income_Category_Freq_table = df['Income Category'].value_counts()
Income_Category_Freq_table
```

```python
dfkeys_IC=Income_Category_Freq_table.keys()
dfkeys_IC
```

```python
dfvals_IC=Income_Category_Freq_table.values
dfvals_IC
```

```python
Income_Category_Freq_table = pd.DataFrame({'Income Category':dfkeys_IC,'Frequency':dfvals_IC})
Income_Category_Freq_table
```

```python
Income_Category_Freq_table ['Percent'] = Income_Category_Freq_table ['Frequency']/Income_Category_Freq_table ['Frequency'].sum()*100
Income_Category_Freq_table
```

```python
import warnings
warnings.filterwarnings('ignore')
plt.figsize = (20,20)
IC = Income_Category_Freq_table['Income Category']
Per=Income_Category_Freq_table['Percent']
plt.xlabel("Frequency Percentage")
plt.ylabel("Income Category")
# Horizontal Bar Plot
plt.barh(IC,Per)
plt.title("Variable 4 - Income Category (Bar Chart)")
plt.show()
```
